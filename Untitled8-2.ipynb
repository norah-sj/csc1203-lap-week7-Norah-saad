```python
try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    result = num1 / num2
except ZeroDivisionError:
    print("Error: Cannot divide by zero!")
except ValueError:
    print("Error: Please enter numbers only!")
else:
    print("Result =", result)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Nora", 19)
p1.age = 19

print("Name:", p1.name)
print("Updated Age:", p1.age)
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

class Manager(Employee):
    def add_bonus(self, bonus):
        self.salary += bonus

    def display(self):
        print("Name:", self.name)
        print("Total Salary:", self.salary)

m1 = Manager("Nora", 5000)
m1.add_bonus(1000)
m1.display()
```

```python
class Student:
    def __init__(self):
        try:
            self.name = input("Enter name: ")
            self.age = int(input("Enter age: "))
            if self.age < 0:
                raise ValueError
        except ValueError:
            print("Error: Age must be a positive number!")
            self.age = None

    def display(self):
        if self.age is not None:
            print("Name:", self.name)
            print("Age:", self.age)

s1 = Student()
s1.display()
```
